```python
%matplotlib inline
from pandas import *
```

```python
df = read_csv('Downloads/atl_salary_data_2015.csv')
del df['name']
```

# Average salaries

> Note we did not clean the data at all.

```python
g = df[['ethnic.origin', 'annual.salary']].groupby('ethnic.origin').describe()['annual.salary'].unstack(-1)

g['mean'].sort_values().plot.bar(grid=True, figsize=(12, 6));
```

## Count of employees sampled

> Note this is on a log scale.

```python
g['count'].sort_values().plot.bar(grid=True, logy=True, figsize=(12, 6));
```
