```python
from bokeh.resources import *
from bokeh.plotting import *
output_notebook(CDN)
```

```python
from pandas import *

triangle = [[0], [0,1], [0,2]], [[0,1], [1], [1,2]], [[0,2], [1,2], [2]]
sub = [[-1,0],[0,1],[1,0]]

df = concat({
    -1: concat({0: DataFrame(sub, range(3))})
}, names=['level', 'triangle', 'node'])

for level in range(2):
    for index in df.loc[level-1].index.get_level_values('triangle').unique():
        df = concat([df, concat({
            level: concat({
                j + index*3: concat({
                    i: df.loc[level-1].loc[index].loc[pairs].mean()
                    for i, pairs in enumerate(t)
                }) for j, t in enumerate(triangle)}).unstack(-1)})])
```

```python
_df = df.copy()
for q in range(4):
    _df = concat([
        _df.add(_) for _ in sub
    ]).div([2, 2])
_df.sample(2)
```

```python
source = _df.pipe(
    lambda df: setattr(df, 'columns', df.columns.astype(str)) or df).pipe(ColumnDataSource)
```

```python
p = figure(output_backend="webgl", responsive=True)
p.circle(*list('01'), source=source)
show(p)
```
