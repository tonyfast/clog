```python
from bokeh.resources import *
from bokeh.models import *
from bokeh.plotting import *
output_notebook(CDN)
```

```python
from pandas import *

triangle = [[0], [0,1], [0,2]], [[0,1], [1], [1,2]], [[0,2], [1,2], [2]]
sub = [[0,0], [.8, .5],[1,0]]


def generate_base():
    df = concat({
        -1: concat({0: DataFrame(sub, range(3))})
    }, names=['level', 'triangle', 'node'])

    for level in range(2):
        for index in df.loc[level-1].index.get_level_values('triangle').unique():
            df = concat([df, concat({
                level: concat({
                    j + index*3: concat({
                        i: df.loc[level-1].loc[index].loc[pairs].mean()
                        for i, pairs in enumerate(t)
                    }) for j, t in enumerate(triangle)}).unstack(-1)})])
    return df
```

```python
def generate(levels=3):
    df=generate_base()
    for q in range(levels):
        df = concat([
            df.add(_) for _ in sub
        ]).div([2, 2])
    return print("There are {} triangles.".format(df.pipe(len)/3)) or df
df = generate()
df.sample(2)
```

```python
source = df.pipe(
    lambda df: setattr(df, 'columns', df.columns.astype(str)) or df).reset_index(['triangle', 'node']).pipe(ColumnDataSource)
```

```python
p = figure(output_backend="webgl", responsive=True)
rotation = 0 
for s in range(10, 90, 5):
    p.circle(
        *list('01'), source=source, fill_alpha=.1, line_alpha=0, size=s,
        fill_color=dict(
            field='node', transform=CustomJSTransform(
                args=dict(source=source,),
                v_func="""return source.data.node.map(function(v,i){return ['red', 'green', 'blue'][v];});""")))
show(p)
```

```python
sub = [[0,0], [.2, 1],[1,0]]
```

```python
df = generate(4)
source = df.pipe(
    lambda df: setattr(df, 'columns', df.columns.astype(str)) or df).reset_index(['triangle', 'node']).pipe(ColumnDataSource)
source.add(list(map("".__getitem__, source.data['node'])), 'text');
```

```python
p = figure(output_backend="webgl", responsive=True)
p.text(
    *list('01'),  source=source, text_font_size='5rem', text_alpha=.6,)
show(p)
```
